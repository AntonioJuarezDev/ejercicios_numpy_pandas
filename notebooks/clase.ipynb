```python
import numpy as np
import pandas as pd
```

```python
users = pd.read_csv('../csv/users_v2.csv')
```

### Basic dataframe syntax
data[**filter(s)**][**column(s)**]

### Obtener el nombre y correo de todos los usuarios.

```python
users[ ['name', 'email'] ]
```

### Obtener el nombre de todos los usuarios cuyo sexo es Femenino.

```python
users.columns
```

```python
users[ users['gender'] == 'female' ]['name']
```

### Obtener el nombre de todos los usuarios cuyo sexo es Masculino y tengan una edad mayuor a 50.

```python
users[ (users['gender'] == 'male') & (users['age'] > 50) ]['name']
```

### Obtener todos los usuarios cuya edad sea mayor a 20. (Fue mayor a 68 en clase)

```python
users[ users['age'] > 68 ]
```

### Obtener el nombre y el pais del usuarios mas joven (Ejercicio visto en clase)

```python
users.sort_values('age').head(1)[ ['name', 'country', 'age'] ]
```

### Obtener el username y correo electrónico de los usuarios cuya edad se encuentre en el rango 10 a 20. (Fue de 50 a 60 en clase)

```python
users[ (users['age'] > 50) & (users['age'] < 60) ][ ['username', 'email', 'age'] ]
```

### Obtener el username de todos los usuarios cuyo correo no terminen con @example.com

```python
users[ ~users['email'].str.endswith('@example.com') ][ ['username', 'email'] ]
```

### Obtener el username de todos los usuarios cuyo país sea Alemania, Finlandia o Canadá.

```python
users[ users['country'].isin( ['Germany', 'Finland', 'Canada'] ) ][ ['username', 'country'] ]
```

### Obtener el nombre y correo electronico de todos los usuarios de sexo Femenino que sean del país Alemania.

```python
users[ (users['gender'] == 'female' ) & ( users['country'] == 'Germany' ) ][ ['name', 'email'] ]
```

### Obtener el promedio de edad de todos los usuarios de sexo Femenido de Canada mayores de 20 años.

```python
users[ (users['gender'] == 'female' ) & ( users['country'] == 'Canada' ) & ( users['age'] > 20 ) ]['age'].mean()
```

### Conocer la cantidad de usuarios que residen en Finlandia.

```python
len(users[ users['country'] == 'Finland' ])
```

### Mostrar en consola la cantidad de hombre y mujeres en el dataframe.

```python
male = len(users[ users['gender'] == 'male' ])
female = len(users[ users['gender'] == 'female' ])

print(f"Hombres: {male}")
print(f"Mujeres: {female}")
```

```python
users.groupby('gender')['gender'].count()
```

### Mostrar en consola el país con más mujeres

```python
users[ users['gender'] == 'female' ].groupby('country')['country'].count().sort_values().tail(1)
```

### Obtener los 3 países con más usuarios.

```python
users.groupby('country')['country'].count().sort_values(ascending = False).head(3)
```
